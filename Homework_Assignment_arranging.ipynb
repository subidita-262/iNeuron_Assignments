```python
import os
import shutil
```

Task 1 : To rename the files and make separate directory for the different assignments

```python
def rename(file):
    """
    This function renames a file by replacing conventional brackets '()' for duplicate names by '_'
    Source file needs to be passed as an argument and the function will rename the file in the directory.
    """
    new = file.replace('(', '_')
    new1 = new.replace(')', '')
    #print(new)
    os.rename(file, new1)

prev_dir = os.getcwd()
try:
    os.mkdir("Arranged_assignments")
except FileExistsError:
    print("Directory exits")
    
os.chdir(prev_dir+'\Assignment_questions')    
for i in os.listdir():
    #print(i)
    if i[-6] == ')':
        rename(i)
    if 'Programming' in i:
        directory = '3'+i.split('_')[0] + '-Basic'
    elif i.count("_") == 2:
        directory = '2'+i.split('_')[0] + '-Python-Advance'
    elif i.count("_") == 1:
        directory = '1'+i.split('_')[0] + '-Python-Basics'
    elif i.count("_") == 0:
        directory = '4Programming-' + i.split('t')[0] + 't-Advanced'
    os.chdir(prev_dir+'\Arranged_assignments')
    try:
        os.mkdir(directory)
    except FileExistsError:
        print("Directory exits")
    os.chdir(prev_dir+'\Assignment_questions')
    newpath = prev_dir+'\Arranged_assignments' + '\\' + directory
    #print(newpath)
    shutil.move(i, newpath)
    #print("done")
        
```

Task 2 : To make the combined file to combine all the questions in one file

```python
from docx import Document
import natsort
os.chdir(prev_dir+'\Arranged_assignments')
resfile = open("combined-file.txt", 'w', encoding="utf-8")
print(os.getcwd())
result_dir = os.getcwd()
print(os.listdir())
for i in os.listdir():
    if i == 'combined-file.txt':
        continue
    os.chdir(os.getcwd() +'\\' + i)
    resfile.write('$$$'+i[1:] + '\n' +'\n')
    files = natsort.natsorted(os.listdir())
    for file in files:
        if file == 'combined-file.txt':
            continue
        #print('file=',file)
        resfile.write('-->'+file + '\n' +'\n')
        document = Document(file)
        for para in document.paragraphs:
            #print(para.text)
            resfile.write(para.text + '\n')
        resfile.write('\n' + '\n')
    #print(sorted(files))
    #print("--------------")
    os.chdir(result_dir)

resfile.close()
    
```
